```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/Training.csv')
```

```python
X = df.drop('prognosis',axis=1)
y = df['prognosis']
```

```python
label = LabelEncoder()
label.fit(y)
Y = label.transform(y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=46)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
Y_train
```

```python
Y_test
```

```python
# creating a dictonary to run all models in a loop
models = {
    "RandomForestClassifier":RandomForestClassifier(n_estimators=100,random_state=46),
    "GradientBoostingClassifier":GradientBoostingClassifier(n_estimators=100,random_state=46),
    "SVC":SVC(kernel='linear'),
    "KNeighborsClassifier":KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB":MultinomialNB()
}
for model_name,model in models.items():
  model.fit(X_train,Y_train)
  Y_pred = model.predict(X_test)
  CM = confusion_matrix(Y_pred,Y_test)
  print(f"Accuracy score of {model_name} is {accuracy_score(Y_pred,Y_test)}")
  print(f"Confusion matrix of {model_name} is ")
  print(np.array2string(CM,separator=','))
```

```python
svc = SVC(kernel='linear')
svc.fit(X_train,Y_train)
Y_pred=svc.predict(X_test)
print(accuracy_score(Y_pred,Y_test))
```

```python
def helper(predicted_disease):
  desc = description[description['Disease']==predicted_disease]['Description']
  desc = " ".join([w for w in desc])
  prec = precautions[precautions['Disease']==predicted_disease][['Precaution_1','Precaution_2','Precaution_3','Precaution_4']]
  prec = [col for col in prec.values]
  med = medications[medications['Disease']==predicted_disease]['Medication']
  med = [med for med in med.values]
  diet = diets[diets['Disease']==predicted_disease]['Diet']
  diet = [diet for diet in diet.values]
  return desc,prec,med,diet
```
